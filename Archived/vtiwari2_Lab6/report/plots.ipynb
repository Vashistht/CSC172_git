```python
def logify(arr):
    arr_temp = arr
    for i in range(len(arr)):
        if arr[i] == 0:
            arr_temp[i] == 0
        else:
            arr_temp[i] == np.log10(arr)
    return arr_temp
```

```python

plt.rcParams["figure.figsize"] = (12,6)


plt.scatter(arraysize, logify(java_def_a), marker = "o",
                        label=r"java_def", alpha=0.5)

plt.scatter(arraysize, logify(bubble_sort_a), marker = "v",
                        label=r"bubble_sort", alpha=0.5)

plt.scatter(arraysize, logify(selection_sort_a), marker = "s",
                        label=r"selection_sort", alpha=0.5)


plt.scatter(arraysize, logify(insertion_sort_a), marker = "P",
                        label=r"insertion_sort", alpha=0.5)

plt.scatter(arraysize, logify(mergesort_a), marker = "*",
                        label=r"merge_sort", alpha=0.5)

plt.scatter(arraysize, logify(quicksort_a), marker = "h",
                        label=r"quick_sort", alpha=0.5)

plt.annotate('Nan', xy =(1000, 0), fontsize=13)
plt.xlabel(r"Length of the array (log-scaled)", fontsize=15)
plt.ylabel("Runtime (log-scaled) [in ms]", fontsize=15)
plt.xscale("log")
plt.title("Run time vs array length for list a", fontsize = 25)
plt.legend(loc="best", fontsize=15)
plt.tight_layout()
plt.savefig("plot_lista", dpi=300)
plt.show()
```
